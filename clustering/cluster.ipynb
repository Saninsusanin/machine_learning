# 1

```r
library(cluster)
data("pluton")
for(i in seq(2, 16, by = 2))
{
  cl <- kmeans(pluton, 3, iter.max = i)
  plot(pluton, col = cl$cluster, main=paste('max=',toString(i)))
  points(cl$centers, col = 1:3, pch = 8, cex=2)
}
```

# 2

```r
cl1 <- cbind(matrix(rnorm(100, mean = 0, sd = 10), ncol = 1),
             matrix(rnorm(100, mean = 0, sd = 0.5), ncol = 1))
cl2 <- cbind(matrix(rnorm(100, mean = 0, sd = 10), ncol = 1),
             matrix(rnorm(100, mean = 5, sd = 0.5), ncol = 1))
cl3 <- cbind(matrix(rnorm(100, mean = 0, sd = 10), ncol = 1),
             matrix(rnorm(100, mean = 10, sd = 0.5), ncol = 1))
df <- rbind(cl1, cl2, cl3)

true_man<-clara(df, 3, metric = "manhattan", stand = TRUE)
true_eucl<-clara(df, 3, metric = "euclidean", stand = TRUE)
not_true_man<-clara(df, 3,metric = "manhattan", stand = FALSE)
not_true_eucl<-clara(df, 3, metric = "euclidean", stand = FALSE)

plot(df, col = true_man$clustering)
plot(df, col = true_eucl$clustering)
plot(df, col = not_true_man$clustering)
plot(df, col = not_true_eucl$clustering)
```
